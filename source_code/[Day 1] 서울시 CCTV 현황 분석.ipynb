```python
import pandas as pd
cctv_seoul = pd.read_excel('../data/01.cctv_in_seoul.xlsx', encoding='UTF-8')
cctv_seoul.head()
# type(cctv_seoul)
```

```python
cctv_seoul.columns
```

```python
cctv_seoul.rename(columns={cctv_seoul.columns[0]: '구별'}, inplace=True)
cctv_seoul.head()
```

```python
pop_seoul = pd.read_excel('../data/01.population_in_seoul.xls',
                          header=2, 
                          usecols='B, D, G, J, N',
                          encoding='utf-8')
pop_seoul.head()
```

```python
pop_seoul.rename(columns={pop_seoul.columns[0]: '구별', pop_seoul.columns[1]: '인구수',pop_seoul.columns[2]: '한국인', pop_seoul.columns[3]: '외국인', pop_seoul.columns[4]: '고령자'}, inplace=True)
pop_seoul.head()
```

```python
cctv_seoul.sort_values(by='소계', ascending=False).head(5)
```

```python
cctv_seoul.sort_values(by='소계', ascending=True).head(5)
```

```python
cctv_seoul['최근 증가율'] = (cctv_seoul['2018년'] \
                        + cctv_seoul['2017년'] \
                        + cctv_seoul['2016년']) \
                        / (cctv_seoul['2015년'] \
                        + cctv_seoul['2014년'] \
                        + cctv_seoul['2013년'] \
                        + cctv_seoul['2012년'] \
                        + cctv_seoul['2011년 이전']) * 100
cctv_seoul
```

```python
pop_seoul.drop([0], inplace=True)
pop_seoul.head()
```

```python
pop_seoul['구별'].unique()
```

```python
pop_seoul[pop_seoul['구별'].isnull()]
```

```python
pop_seoul.head()
```

```python
pop_seoul["외국인 비율"] = (pop_seoul['외국인']/pop_seoul['인구수'])*100
pop_seoul["고령자 비율"] = (pop_seoul['고령자']/pop_seoul['인구수'])*100
pop_seoul.head()
```

```python
pop_seoul.sort_values(by=['고령자 비율', '인구수'] , ascending=False).head(2)
```

```python
pop_seoul.sort_values(by='고령자 비율', ascending=True).head(1)
```

```python
pop_seoul.sort_values(by='외국인 비율', ascending=False).head(1)
```

```python
pop_seoul.sort_values(by='외국인 비율', ascending=True).head(1)
```

```python
cctv_seoul['구별'] = cctv_seoul['구별'].replace(to_replace=" ", value="", regex = True)
cctv_seoul.head()
```

```python
data_result = pd.merge(cctv_seoul, pop_seoul, on='구별')
data_result.head()
```

```python
del data_result['2011년 이전']
del data_result['2012년']
del data_result['2013년']
del data_result['2014년']
del data_result['2015년']
del data_result['2016년']
del data_result['2017년']
del data_result['2018년']
data_result.head()
```

```python
data_result.set_index('구별', inplace=True)
data_result.head()
```

```python
import numpy as np
np.corrcoef(data_result['고령자 비율'], data_result['소계'])
```
